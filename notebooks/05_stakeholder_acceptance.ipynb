## Proyek: Segmentasi Pelanggan Online Retail Berbasis RFM (Spark)

Stage terakhir dalam **Team Data Science Process (TDSP)** ini berfokus pada **validasi hasil bersama stakeholder**,
penerimaan bisnis (*customer acceptance*), serta penyusunan **langkah lanjutan** dari proyek analitik.

## 5.1 Tujuan Customer Acceptance

Tujuan utama tahap ini adalah memastikan bahwa:

- Hasil analisis dapat dipahami oleh stakeholder non-teknis
- Segmentasi pelanggan relevan dengan kondisi bisnis nyata
- Insight yang dihasilkan dapat digunakan untuk pengambilan keputusan
- Risiko dan keterbatasan model dipahami dengan baik


## 5.2 Validasi Hasil Segmentasi

Validasi dilakukan secara **kualitatif dan kuantitatif** bersama stakeholder bisnis.

Aspek validasi yang diperhatikan:
- Apakah karakteristik tiap cluster masuk akal secara bisnis?
- Apakah proporsi pelanggan dan kontribusi revenue realistis?
- Apakah segmentasi membantu membedakan strategi pemasaran?

Pada tahap ini, feedback stakeholder sangat penting untuk memastikan adopsi hasil analitik.

## 5.3 Risiko & Keterbatasan Model

Beberapa keterbatasan dari pendekatan yang digunakan:

- Model RFM bersifat **deskriptif**, bukan prediktif
- Tidak mempertimbangkan faktor non-transaksional (demografi, web behavior)
- Sensitif terhadap periode waktu analisis
- Clustering dapat berubah jika data baru ditambahkan

Pemahaman atas keterbatasan ini penting agar hasil tidak disalahartikan.

## 5.4 Rencana Tindak Lanjut (Next Steps)

Setelah proyek ini diterima, beberapa pengembangan lanjutan yang direkomendasikan:

1. **Association Rule Mining**  
   Mengidentifikasi produk yang sering dibeli bersama oleh tiap segmen pelanggan.

2. **Customer Lifecycle Value (CLV)**  
   Mengembangkan model prediktif untuk mengestimasi nilai pelanggan jangka panjang.

3. **Churn Prediction**  
   Mengidentifikasi pelanggan berisiko churn berdasarkan histori transaksi.

4. **Integrasi Dashboard BI**  
   Visualisasi segmentasi dan KPI pelanggan secara real-time.


## 5.5 Kriteria Keberhasilan Proyek

Proyek ini dianggap sukses apabila:

- Stakeholder menerima dan memahami hasil segmentasi
- Segmentasi digunakan sebagai dasar strategi marketing
- Terdapat rencana konkret untuk analitik lanjutan
- Proyek dapat direplikasi atau diskalakan ke periode data lain


## Ringkasan Akhir TDSP

Melalui lima stage TDSP, proyek ini telah:

1. Mendefinisikan permasalahan bisnis dengan jelas
2. Memahami dan menyiapkan data transaksi menggunakan Spark
3. Melakukan modeling RFM dan segmentasi pelanggan
4. Menerjemahkan hasil analitik ke insight bisnis
5. Memvalidasi hasil dan menyusun roadmap lanjutan

Pendekatan ini memastikan bahwa analitik tidak hanya **teknis benar**,
tetapi juga **bernilai bagi bisnis**.
